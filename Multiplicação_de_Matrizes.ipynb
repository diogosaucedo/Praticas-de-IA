# Multiplicação de matrizes

## Importando lib



```python
%tensorflow_version 2.x
import tensorflow as tf
```

## Exercicios
A)

```python
matriz_a = tf.constant([[1,0],[2,1]])
matriz_b = tf.constant([[0,2],[6,1]])
```

```python
if matriz_a.shape[1] == matriz_b.shape[0]:
  print("Mulplicação Valida!")
  resultado = tf.matmul((matriz_a * 2),( matriz_b * 3))
  print(resultado)
else:
  print("As Matrizes não podem ser multiplicadas")
```
